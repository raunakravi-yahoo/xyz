```python
df3 = data['3_years'].copy()
df3.info(); display(df3)

df3 = df3[['CDMS ID'] + df3.columns[7:].to_list()]

COI3 = ['Count of Fillings ', 'Count of Grants',
       'Patent Application Growth 3-year CAGR',
       'Patent Grants Growth 3-year CAGR', 
       'Share of Self-Cited Patents', 
       'Average Forward Citations ', 'Generality Index', 'Tail Innovations',
       'Global Technological Collaboration',
       'Number of Companies citing your patents ', 'Originality Index',
       'Technology Cycle Time',
       'Share of Hiring Activity in Disruptive Themes',
       'Share of Patents Granted in Disruptive Themes']
for col in COI3:
    # Calculate min-max normalized values for selected columns
    min_value = df3[col].min()
    max_value = df3[col].max()

    # Normalized values calculation adjusted
    normalized_values_rounded = 0.01 + (df3[col] - min_value) * (1 - 0.01) / (max_value - min_value)
    #normalized_values_rounded = normalized_values.round(2)

    # Add the calculated values to the original DataFrame
    new_col_name = 'Normalized ' + col
    df3[new_col_name] = normalized_values_rounded


COI3_2 = ['Innovation Sentiment']


for col in COI3_2:
    # Calculate min-max normalized values for selected columns
    min_value = df3[col].min()
    max_value = df3[col].max()

    # Adjusted normalized values calculation with condition
    normalized_values_rounded = np.where(
        df3[col].isna(),  # Condition to check if the value is NaN (blank)
        0.01,  # Value if condition is true
        0.01 + (df3[col] - min_value) * (1 - 0.01) / (max_value - min_value)  # Value if condition is false
    )
    
    #normalized_values_rounded = np.round(normalized_values, 2)

    # Add the calculated values to the original DataFrame
    new_col_name = 'Normalized ' + col
    df3[new_col_name] = normalized_values_rounded
    
    
COI3_3 = ['Patents Portfolio Validity']
for col in COI3_3:
    # Calculate min-max normalized values for selected columns
    min_value = df3[col].min()
    max_value = df3[col].max()

    # Step 1: Set to 0 if df3[col] < 0
    intermediate_values = df3[col].apply(lambda x: 0 if x < 0 else x)

    # Step 2: Calculate using the formula on intermediate values
    #normalized_values = 0.01 + (df3[col] - min_value) * (1 - 0.01) / (max_value - min_value)
    normalized_values_rounded = intermediate_values.apply(
        lambda x: 0.01 + ((x - min_value) * (1 - 0.01)) / (max_value - min_value)
        #if x != 0 else 0
    )
    
    #normalized_values_rounded = normalized_values.round(2)

    # Add the calculated values to the original DataFrame
    new_col_name = 'Normalized ' + col
    df3[new_col_name] = normalized_values_rounded

COI3_4 = ['Volume of CVC deals done ','Share of Deals done in Disruptive Themes']
for col in COI3_4:
    # Calculate min-max normalized values for selected columns
    min_value = df3[col].min()
    max_value = df3[col].max()

    # Normalized values calculation adjusted with condition
    normalized_values_rounded = df3[col].apply(lambda x: 0.01 if x < 0 else 0.01 + (x - min_value) * (1 - 0.01) / (max_value - min_value))
    #normalized_values_rounded = normalized_values.round(2)

    # Add the calculated values to the original DataFrame
    new_col_name = 'Normalized ' + col
    df3[new_col_name] = normalized_values_rounded

df3 = df3[['CDMS ID','Normalized Count of Fillings ', 'Normalized Count of Grants',
       'Normalized Patent Application Growth 3-year CAGR',
       'Normalized Patent Grants Growth 3-year CAGR',
       'Normalized Share of Self-Cited Patents',
       'Normalized Average Forward Citations ', 'Normalized Generality Index',
       'Normalized Tail Innovations',
       'Normalized Global Technological Collaboration',
       'Normalized Number of Companies citing your patents ',
       'Normalized Originality Index', 'Normalized Technology Cycle Time',
       'Normalized Share of Hiring Activity in Disruptive Themes',
       'Normalized Share of Patents Granted in Disruptive Themes',
        'Normalized Innovation Sentiment',
       'Normalized Patents Portfolio Validity',
       'Normalized Volume of CVC deals done ',
       'Normalized Share of Deals done in Disruptive Themes']]

### Removing Normalized from column names
df3.columns = df3.columns.str.replace('Normalized ', '')
```

```python
## Logarithmic
def log_exp(value):
    try:
        return np.log(value)
    except (TypeError, ValueError, ZeroDivisionError, OverflowError):
        return ""

Log_columns = df3.columns[1:]
for col in Log_columns:
    normalized_values_rounded = df3[col].apply(log_exp)
    #normalized_values_rounded = normalized_values.round(2)

    # Add the calculated values to the original DataFrame
    new_col_name = 'log ' + col
    df3[new_col_name] = normalized_values_rounded
    
df3 = df3[['CDMS ID','log Count of Fillings ','log Count of Grants', 'log Patent Application Growth 3-year CAGR',
       'log Patent Grants Growth 3-year CAGR',
       'log Share of Self-Cited Patents', 'log Average Forward Citations ',
       'log Generality Index', 'log Tail Innovations',
       'log Global Technological Collaboration',
       'log Number of Companies citing your patents ', 'log Originality Index',
       'log Technology Cycle Time',
       'log Share of Hiring Activity in Disruptive Themes',
       'log Share of Patents Granted in Disruptive Themes',
       'log Innovation Sentiment',
       'log Patents Portfolio Validity', 'log Volume of CVC deals done ',
       'log Share of Deals done in Disruptive Themes']]

# Remove log from names
df3.columns = df3.columns.str.replace('log ', '')

# Z- score
Zscore_columns3 = df3.columns[1:]
for col in Zscore_columns3:
    # Calculate the mean and standard deviation of the column
    mean = df3[col].mean()
    std_dev = df3[col].std()

    # Calculate the Z-score for each value in the column
    normalized_values_rounded = (df3[col] - mean) / std_dev
    #normalized_values_rounded = normalized_values.round(2)
    

    # Add the calculated values to the original DataFrame
    new_col_name = 'Zscore ' + col
    df3[new_col_name] = normalized_values_rounded

    
df3 = df3[['CDMS ID','Zscore Count of Fillings ','Zscore Count of Grants',
       'Zscore Patent Application Growth 3-year CAGR',
       'Zscore Patent Grants Growth 3-year CAGR',
       'Zscore Share of Self-Cited Patents',
       'Zscore Average Forward Citations ', 'Zscore Generality Index',
       'Zscore Tail Innovations', 'Zscore Global Technological Collaboration',
       'Zscore Number of Companies citing your patents ',
       'Zscore Originality Index', 'Zscore Technology Cycle Time',
       'Zscore Share of Hiring Activity in Disruptive Themes',
       'Zscore Share of Patents Granted in Disruptive Themes',
       'Zscore Innovation Sentiment',
       'Zscore Patents Portfolio Validity', 'Zscore Volume of CVC deals done ',
       'Zscore Share of Deals done in Disruptive Themes']]

df3.columns = df3.columns.str.replace('Zscore ', '')
```

```python
# Handle Outliers
Zscore_out3 = ['Count of Fillings ','Count of Grants', 'Patent Application Growth 3-year CAGR',
       'Patent Grants Growth 3-year CAGR', 'Share of Self-Cited Patents',
       'Average Forward Citations ', 'Generality Index', 'Tail Innovations',
       'Global Technological Collaboration',
       'Number of Companies citing your patents ', 'Originality Index',
       'Technology Cycle Time',
       'Share of Hiring Activity in Disruptive Themes',
       'Share of Patents Granted in Disruptive Themes', 
       'Innovation Sentiment', 'Patents Portfolio Validity',
       'Volume of CVC deals done ',
       'Share of Deals done in Disruptive Themes']
for col in Zscore_out3:
    
    # Identify outliers with Z-scores > 3.5
    outliers = df3[col] > 3.5

    # Replace outlier values with the next value just greater than 3.5
    next_greater_value = df3.loc[df3[col] > 3.5, col].min()
    df3.loc[outliers, col] = next_greater_value


for col in Zscore_out3:
    
    # Identify outliers with Z-scores < -3.5
    outliers1 = df3[col] < -3.5

    # Replace outlier values with the next value just greater than -3.5
    next_greater_value = df3.loc[df3[col] < -3.5, col].max()
    df3.loc[outliers1, col] = next_greater_value
```
